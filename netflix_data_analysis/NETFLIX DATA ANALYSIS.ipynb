# Q2 In which year highest no of tv shows and movies were released show with line chart ?

```python
df.dtypes
```

```python
df['Date_N'] = pd.to_datetime(df['release_year'], format='%Y')
df
```

```python
df['Date_N'].dt.year.value_counts()  #it counts the occurance of all individual years in date column
```

```python
year_counts = df['release_year'].value_counts().sort_index()

year_counts.plot(kind='line', marker='o', figsize=(12,6))
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.title("Netflix Content Release Trend Over Years")
plt.show()
```
